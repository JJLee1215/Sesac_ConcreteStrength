```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# train.csv, test.csv, sample_submission.csv 3개 파일 로드
# 단 id 컬럼은 인덱스로 지정 (train, test만)
import os
# csv -> df 로드
path        = '/content/drive/MyDrive/SeSAC/share/머신러닝/Dataset/Concrete Dataset'
train       = pd.read_csv( os.path.join( path, 'train.csv'),  index_col='id')
test        = pd.read_csv( os.path.join( path, 'test.csv'),   index_col='id')
submssion   = pd.read_csv( os.path.join( path, 'sample_submission.csv') )

# train의 상위값 2개 확인
train.head(2)
```

```python
# train의 컬럼명 확인
'''
   - 'CementComponent', 'BlastFurnaceSlag', 'FlyAshComponent',
       'WaterComponent', 'SuperplasticizerComponent',
       'CoarseAggregateComponent', 'FineAggregateComponent', 'AgeInDays'
   - 'Strength' : target
'''
train.columns
```

```python
# 데이터 볼륨
train.shape, test.shape, submssion.shape
# test의 피처수가 train보다 작음 => target 부분이 없다
# test데이터와 submssion의 데이터수가 일치 => 제출할때 예측한 데이터
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 데이터 준비
X = train.drop( ['Strength'], axis=1 ) #(?, 8)
y = train['Strength'] #(?,)
X.shape, y.shape

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# 알고리즘 생성
lr = LinearRegression()  # OLS 기반
# 학습
lr.fit( X_train, y_train )
# 예측
y_pred = lr.predict( X_test )
y_pred
```

```python
# 평가
# mse
mse = mean_squared_error( y_test, y_pred )
# rmse
rmse = np.sqrt( mse )

# 좋은 점수는 아님, 손실은 0에 가까울수록, r2는 1에 가까울수록 좋은 모델판단
mse, rmse, r2_score( y_test, y_pred )
```

```python
# 최적화(생략)

# 회귀계수
# y절편(혹은 w0), w1x1 + w2x2 + ... (w1, w2, w3, .... : 피처별 회귀계수)
'''
    다중 선형 회귀식
    콘크리트 강도 = 47.327676022236815 +
                    0.02515378 * CementComponent + .... + 0.08894719 * AgeInDays
'''
lr.intercept_, lr.coef_
```

```python
# 정확도 -> 20% 성능
lr.score( X_test, y_test )
```
